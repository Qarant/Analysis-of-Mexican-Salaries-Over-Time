```python
import pandas as pd
import numpy as np
pd.set_option('future.no_silent_downcasting', True)
```

```python
#Leemos nuestro excel y lo exploramos para poder observar que limpieza de datos necesitamos hacer
df = pd.read_excel("csv/00_raw_data_salaries_mexico_INEGI.xlsx", sheet_name="Hoja 2")
df.head(5)
```

### Podemos observar lo siguiente:
* Formato de información horizontal, este formato no es tan común por lo que debemos tener cuidado al tratar encabezados horizontales
* No posee todos los encabezados ya que en nuestro archivo original son celdas combinadas
* La primer fila la podemos combinar con los años para poder realizar un nuevo encabezado
* Los estados de la república estan en formato horizontal y al estar combinados no los lee de correctamente, por lo que tenemos ciertas celdas vacías 

```python
#En primer lugar elimino la primer fila
df.drop([0], axis=0, inplace=True)
```

```python
#Verifico que se haya eliminado correctamente
df.head()
```

```python
#Realizo un algoritmo para crear nuevos encabezados
años = range (2005, 2024)
trimestre = ["Trimestre I","Trimestre II","Trimestre III","Trimestre IV"]
encabezados = ["Entidad Federativa","Sexo"]

for i in range(len(años)):
    for j in range(len(trimestre)):
        encabezados.append(str(años[i])+" "+trimestre[j])
```

```python
#Asignamos los nuevos encabezados al dataFrame
df.columns = encabezados
df.head()
```

```python
# Ahora exploremos los datos
df.info()
```

```python
# Vamos a quitar los NaN, hay ocasiones en los que los "NaN" no son valores nulos sino cadenas de texto con este valor
# Por lo que primero transformo esas posibles cadenas de texto a un verdadero NaN
df["Entidad Federativa"] = df["Entidad Federativa"].replace("NaN", np.nan)
df.head()
```

```python
#En este caso las celdas vacías las lleno con el metodo forward fill(ffill) y verificamos 
df["Entidad Federativa"] = df["Entidad Federativa"].ffill()
df.tail(5)
```

```python
# Guardo en un archivo diferente para tener seguimiento de los datos
df.to_csv("csv/01_data_cleaning_salaries_mexico_INEGI.csv", sep=";", index=True, encoding="utf-8")
```
